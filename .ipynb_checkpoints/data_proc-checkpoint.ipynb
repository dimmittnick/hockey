```python
df_skaters = data_proc.read_data("~/Desktop/dfs/hockey/", skater=True)
```

```python
df_skaters.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_skaters.fillna(0, inplace=True)
```

```python
df_skaters['blocked'] = 0
```

```python
df_skaters.sort_values(by=['playerId', 'gameDate'], inplace=True)
```

```python
df_skaters = data_proc.add_fantasy(df_skaters, 'fantasy_points')
```

```python
feature_list = ['assists', 'goals', 'plusMinus', 'points', 'fantasy_points', 'shootingPct', 'shots']
```

```python
for feature in feature_list:
    df_skaters[f'{feature}_ma6'] = data_proc.moving_average(df_skaters, feature, 'playerId', 6)
    df_skaters[f'{feature}_ma6'] = df_skaters[f'{feature}_ma6'].shift(1)
```

```python
for feature in feature_list:
    df_skaters[f'{feature}_ma7'] = data_proc.moving_average(df_skaters, feature, 'playerId', 7)
    df_skaters[f'{feature}_ma7'] = df_skaters[f'{feature}_ma7'].shift(1)
    
for feature in feature_list:
    df_skaters[f'{feature}_ma3'] = data_proc.moving_average(df_skaters, feature, 'playerId', 3)
    df_skaters[f'{feature}_ma3'] = df_skaters[f'{feature}_ma3'].shift(1)
    
for feature in feature_list:
    df_skaters[f'{feature}_last_game'] = df_skaters[feature].shift(1)
    
for feature in feature_list:
    df_skaters[f'{feature}_ma10'] = data_proc.moving_average(df_skaters, feature, 'playerId', 10)
    df_skaters[f'{feature}_ma10'] = df_skaters[f'{feature}_ma10'].shift(1)

for feature in feature_list:
    df_skaters[f'{feature}_ma14'] = data_proc.moving_average(df_skaters, feature, 'playerId', 14)
    df_skaters[f'{feature}_ma14'] = df_skaters[f'{feature}_ma14'].shift(1)
```
